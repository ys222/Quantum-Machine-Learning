## Section 12.2: Basic Amplitude Amplification

```python
# Listing 12.1: Equal superposition of two states
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram, plot_bloch_multivector, plot_state_qsphere

qc = QuantumCircuit(1)
qc.h(0)

# execute the qc
results = execute(qc,Aer.get_backend('statevector_simulator')).result()
plot_histogram(results.get_counts())
```

```python
# Listing 12.2: HZH‐circuit
# prepare the circuit
qc = QuantumCircuit(1)
qc.h(0)
qc.z(0)
qc.h(0)

# execute the qc
results = execute(qc,Aer.get_backend('statevector_simulator')).result()
plot_histogram(results.get_counts())
```

```python
# Listing 12.3: HIH‐circuit
# prepare the circuit
qc = QuantumCircuit(1)
qc.h(0)
qc.i(0)
qc.h(0)

# execute the qc
results = execute(qc,Aer.get_backend('statevector_simulator')).result()
plot_histogram(results.get_counts())
```

## Section 12.3: Two-Qubit Amplification

```python
# Listing 12.4: Equal superposition of four states
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram, plot_bloch_multivector, plot_state_qsphere

qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)

# execute the qc
results = execute(qc,Aer.get_backend('statevector_simulator')).result()
plot_histogram(results.get_counts())
```

```python
# Listing 12.5: Two‐qubit Grover searching |10>
qc = QuantumCircuit(2)

qc.h(0)
qc.h(1)
qc.barrier()

qc.x(0)
qc.cz(0, 1)
qc.x(0)

qc.barrier()
qc.h(0)
qc.h(1)

qc.x(0)
qc.x(1)
qc.cz(0,1)
qc.x(0)
qc.x(1)

qc.barrier()
qc.h(0)
qc.h(1)

# execute the qc
results = execute(qc,Aer.get_backend('statevector_simulator')).result()
plot_histogram(results.get_counts())
```

```python
# Additional Listing: Two-qubit Grover searching |01>
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.barrier()
qc.x(1)
qc.cz(0, 1)
qc.x(1)
qc.barrier()
qc.h(0)
qc.h(1)

qc.x(0)
qc.x(1)
qc.cz(0,1)
qc.x(0)
qc.x(1)
qc.h(0)
qc.h(1)
# execute the qc
results = execute(qc,Aer.get_backend('statevector_simulator')).result()
plot_histogram(results.get_counts())
```

```python
# Additional Listing: The full circuit to amplify |00>
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.barrier()

qc.x(0)
qc.x(1)
qc.cz(0, 1)

qc.x(0)
qc.x(1)

qc.barrier()
qc.h(0)
qc.h(1)

qc.x(0)
qc.x(1)
qc.cz(0,1)
qc.x(0)
qc.x(1)
qc.barrier()
qc.h(0)
qc.h(1)
# execute the qc
results = execute(qc,Aer.get_backend('statevector_simulator')).result()
plot_histogram(results.get_counts())
```
